###Data Wrangling

````python
from sklearn.preprocessing import StandardScaler
````

````python
#Code to make the dataset analysis ready

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Drop columns with excessive missing values (>70%)
drop_cols = ["connected_handling_time", "customer_city", "product_category", "item_price", "order_date_time"]
existing_cols = [col for col in drop_cols if col in df.columns]  # Filter for existing columns
df.drop(columns=existing_cols, inplace=True)  # Drop only existing columns

# Fill categorical missing values with "Unknown"
df.fillna({"customer_remarks": "Unknown", "order_id": "Unknown"}, inplace=True)

# Convert categorical variables to category dtype
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].astype("category")

# Remove duplicate records
df.drop_duplicates(inplace=True)

# Handle outliers in numerical columns using IQR capping
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df[col] = np.clip(df[col], lower_bound, upper_bound)

# Normalize numerical columns
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Save the cleaned dataset
df.to_csv("Cleaned_Customer_Support_Data.csv", index=False)

print("Data Wrangling completed. Cleaned dataset saved successfully!")
````

###What all manipulations have you done and insights you found

I performed the following data wrangling and manipulations to make the dataset analysis-ready:

1. Data Cleaning & Preprocessing
   - Standardized Column Names, Removed
     spaces and converted to lowercase.
   - Handled Missing Values,

     - Dropped columns with more than 70% missing values.
     - Filled categorical NaNs with "Unknown".
2. Outlier Handling & Normalization
   - Applied IQR-Based Capping, Handled extreme outliers in numerical columns using the Interquartile Range (IQR) method.
   - Standardized Numerical Features â†’ Used StandardScaler() to normalize numerical columns for better comparison.
