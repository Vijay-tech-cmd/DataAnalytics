## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplfinance as mpf

# Load the dataset
file_path = "/content/data_YesBank_StockPrices.csv"
df = pd.read_csv(file_path)

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')

# Sort the dataframe by date
df = df.sort_values(by='Date')

# Set the Date column as index for candlestick chart
df.set_index('Date', inplace=True)

# Rename columns to match mplfinance requirements
df = df[['Open', 'High', 'Low', 'Close']]

# Plot the line chart
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], marker='o', linestyle='-', color='b', label='Close Price')

# Formatting the chart
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Yes Bank Stock Closing Prices Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.grid()


# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Close'], bins=20, color='g', edgecolor='black', alpha=0.7)

# Formatting the histogram
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Distribution of Yes Bank Closing Prices')
plt.grid(axis='y', linestyle='--', alpha=0.7)


# Plot the box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Close'], vert=True, patch_artist=True, boxprops=dict(facecolor='skyblue'))

# Formatting the box plot
plt.ylabel('Closing Price')
plt.title('Box Plot of Yes Bank Closing Prices')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Plot the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(df.index, df['Close'], color='red', alpha=0.6)

# Formatting the scatter plot
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Scatter Plot of Yes Bank Closing Prices Over Time')
plt.xticks(rotation=45)
plt.grid()


# Plot the candlestick chart
mpf.plot(df, type='candle', style='charles', figsize=(12, 6), title='Yes Bank Stock Candlestick Chart', ylabel='Price')

# Calculate moving averages
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()

# Plot moving average chart
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], color='blue', label='Closing Price')
plt.plot(df.index, df['SMA_50'], color='red', linestyle='dashed', label='50-Day SMA')
plt.plot(df.index, df['SMA_200'], color='green', linestyle='dashed', label='200-Day SMA')

# Formatting the moving average chart
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Yes Bank Stock Moving Averages')
plt.legend()
plt.grid()




# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Formatting the heatmap
plt.title('Correlation Heatmap of Yes Bank Stock Prices')



# Plot the pair plot
sns.pairplot(df, diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Yes Bank Stock Prices', y=1.02)

````
