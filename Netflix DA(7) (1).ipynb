## 3. ***Data Wrangling***

### Data Wrangling Code

````python
netflix_df = df.copy()
````

````python
# Ensure 'date_added' is in datetime format
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], errors='coerce')

# Now extract year and month
netflix_df['year_added'] = netflix_df['date_added'].dt.year
netflix_df['month_added'] = netflix_df['date_added'].dt.month
````

````python
# Data Wrangling: Cleaning and Preparing the Netflix Dataset

# Step 1: Handle missing values
# Fill missing 'director', 'cast', 'country' with 'Unknown'
netflix_df['director'].fillna('Unknown', inplace=True)
netflix_df['cast'].fillna('Unknown', inplace=True)
netflix_df['country'].fillna('Unknown', inplace=True)
netflix_df['rating'].fillna('Not Rated', inplace=True)

# Step 2: Convert 'date_added' to datetime (already done previously)
# Extract 'year_added' and 'month_added' for analysis
netflix_df['year_added'] = netflix_df['date_added'].dt.year
netflix_df['month_added'] = netflix_df['date_added'].dt.month

# Step 3: Clean 'duration' column


# Separate numeric part and unit from 'duration'
netflix_df['duration_int'] = netflix_df['duration'].str.extract('(\d+)').astype(float)
netflix_df['duration_type'] = netflix_df['duration'].str.extract('([a-zA-Z]+)')

# Step 4: Handle multiple genres in 'listed_in'
# For genre analysis, we can split and explode the genres later

# Step 5: Normalize column names (optional, for consistency)
netflix_df.columns = [col.strip().lower().replace(' ', '_') for col in netflix_df.columns]

# Step 6: Remove duplicates if any
netflix_df.drop_duplicates(inplace=True)

````

### What all manipulations have you done and insights you found?

In this project, several data cleaning and transformation steps were performed to prepare the Netflix Movies and TV Shows dataset for analysis. First, the dataset was loaded and inspected for missing values. Columns such as cast, director, and country contained missing entries, which were retained since they do not critically affect the analysis. The date_added column was converted to datetime format to extract the year_added, which helped in analyzing content addition trends over time.

The duration column, which contained mixed values (e.g., "90 min" for movies and "2 Seasons" for TV shows), was split into two separate columns: duration_int (numeric value) and duration_type (minutes or seasons). To better analyze the listed_in (genre) field—which often contains multiple genres separated by commas—it was split and expanded using the explode() function, allowing for more accurate genre-level analysis.

Using these cleaned and processed fields, we generated several visualizations to uncover meaningful insights:

Type Distribution: The majority of the content on Netflix is movies. This indicates a strong focus on film content, although the platform could benefit from increasing high-quality TV shows to improve long-term viewer retention.

Content Ratings: The most common ratings are TV-MA and PG-13, showing that Netflix targets more mature audiences. However, there is relatively little family or kid-friendly content, which could be an area for growth.

Release Year Trends: Most of the content was released after the year 2000, with a significant spike in releases between 2015 and 2020. This suggests a focus on modern content, though there may be opportunities in licensing classic films and series.

Genre Insights: The most frequent genres include Dramas, Comedies, and International content, reflecting a global and emotional storytelling strategy. However, genres like documentaries, animation, and sci-fi are comparatively underrepresented.

Duration vs. Rating: Mature-rated content (e.g., TV-MA, R) tends to be longer in duration, possibly reflecting more complex storytelling or series-style films. This insight helps align content length with audience expectations.

Yearly Additions: Netflix’s yearly content additions peaked in 2019–2020 and declined slightly afterward, possibly due to licensing shifts or pandemic-related production issues.

These insights can help Netflix refine its content acquisition and production strategies, enhance user satisfaction, and identify under-served audience segments. Overall, the analysis supports data-driven decision-making for business growth while also highlighting areas where over-saturation or under-delivery could lead to negative impact if not addressed.

#### Chart - 1

````python
# Convert 'date_added' to datetime
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], errors='coerce')

# Extract release year and clean text fields
netflix_df['release_year'] = pd.to_numeric(netflix_df['release_year'], errors='coerce')
for col in ['type', 'title', 'director', 'country', 'rating']:
    netflix_df[col] = netflix_df[col].astype(str).str.strip()

# Set visualization style
sns.set(style="whitegrid")

# Plot: Content Released by Year
plt.figure(figsize=(12, 6))
sns.countplot(
    x='release_year',
    data=netflix_df,
    order=netflix_df['release_year'].value_counts().sort_index().index,
    palette='Blues_r'
)
plt.title('Netflix Content Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

#### Chart - 6 - Correlation Heatmap

````python
# Preprocess: extract duration in numbers and release/added year
netflix_df['duration'] = netflix_df['duration'].astype(str).str.strip()
netflix_df['duration_int'] = netflix_df['duration'].str.extract('(\d+)').astype(float)
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], errors='coerce')
netflix_df['added_year'] = netflix_df['date_added'].dt.year
netflix_df['release_year'] = pd.to_numeric(netflix_df['release_year'], errors='coerce')

# Select numeric columns
numeric_df = netflix_df[['duration_int', 'release_year', 'added_year']]

# Compute correlation
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Netflix Numeric Features')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

- A correlation heatmap shows the strength and direction of linear relationships between numerical features.

- It’s crucial for identifying which variables move together (positive correlation) or in opposite directions (negative correlation).

##### 2. What is/are the insight(s) found from the chart?

- Moderate correlation may be seen between release_year and added_year, suggesting that newer content tends to be added more quickly.

- Duration generally shows low correlation with years, meaning Netflix’s movie/show length doesn’t trend strongly over time.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

#### Chart - 7 - Pair Plot
